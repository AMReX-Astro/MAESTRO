```python
target_hs = None
for hs in r12030_175H._out._hotspots:
    print(hs.timestamp[0])
    #if hs.timestamp[0] == 4231:
    #if hs.timestamp[0] == 4361:
    if hs.timestamp[0] == 5339:
        target_hs = hs
        
```

```python
len(target_hs._igarrays)
```

```python
theta = []
phi = []
r = []
itheta = []
iphi = []
ir = []

for arr in target_hs._igarrays:
    theta_sub = [] 
    phi_sub = []
    r_sub = []
    for hs in arr:
        theta_sub.append(hs.loc[1][1])
        phi_sub.append(hs.loc[1][2])
        r_sub.append(hs.loc[1][0])
        
    theta.append(theta_sub)
    phi.append(phi_sub)
    r.append(r_sub)

for Ig in target_hs._Ig_arr:
    if Ig.M/1.e24 > 1.5:
        itheta.append(Ig.theta)
        iphi.append(Ig.phi)
        ir.append(Ig.r)

    
    
theta = np.array(theta)
phi = np.array(phi)
r = np.array(r)
itheta = np.array(itheta)
iphi = np.array(iphi)
ir = np.array(ir)
```
