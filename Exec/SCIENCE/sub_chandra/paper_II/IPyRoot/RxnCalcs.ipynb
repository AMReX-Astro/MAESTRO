It's pointed out in Woosley and Kasen 2011 that the following are important reactions to include when investigating the explosive dynamics of the sub-Chandrasekhar mass type Ia progenitor scenario:

$^{14}\mathrm{N}(e^-, \nu)^{14}\mathrm{C}(\alpha, \gamma)^{18}\mathrm{O}$,  
$^{12}\mathrm{C}(p, \gamma)^{13}\mathrm{N}(\alpha, p)^{16}\mathrm{O}$.

Here I discuss these reactions and demonstrate why we are ignoring them in the sub-Ch II paper.

### $^{14}\mathrm{N}(e^-, \nu)^{14}\mathrm{C}(\alpha, \gamma)^{18}\mathrm{O}$

This reaction is mainly crucial in the lead-up to 3$\alpha$ burning, but becomes unimportant to energetics once 3$\alpha$ burning's ignited.  This reaction serves to reduce the density at which 3$\alpha$ is first ignited by dramatically increasing the temperature at a given density.  Without NCO more accretion is required until the density is high enough that 3$\alpha$ can be ignited at lower temperatures.  For discussion of this see Hashimoto, M.-A., Nomoto, K.-I., Arai, K., & Kaminisi, K. 1986, ApJ, 307, 687, especially Figure 5.

Furthermore, the impact of NCO triggering 3$\alpha$ is most pronounced at $\rho > 10^{6.5}~\rm{g}~\rm{cm}^{-3}$, which is rarely the case at the sites of burning in our sub-Chandrasekhar models.  Below this most of the $^{14}\mathrm{C}$ beta decays back to  $^{14}\mathrm{N}$ faster than it's created.

Since we set our temperture well into the 3$\alpha$ burning regime and aren't doing long-term evolution/accretion we can neglect this rate.  At the thermodynamic conditions we're concerned with ($T \sim 200$ MK, 
$\rho \sim 10^6~\rm{g}~\rm{cm}^{-3}$) the NCO energy generation is several orders of magnitude less than 3$\alpha$'s.  However, it's only about half an order of magnitude less than CAGO's ($^{12}\mathrm{C}(\alpha, \gamma)^{16}\mathrm{O}$) contribution (assuming enough $^{14}\rm{C}$ has been produced via $e^-$ capture to be able to release much heat), which we _do_ include in our models.  It would be interesting to include it in the following detailed study, but for the broad study its energetics are sufficiently low to neglect.

### $^{12}\mathrm{C}(p, \gamma)^{13}\mathrm{N}(\alpha, p)^{16}\mathrm{O}$

This reaction acts as a CAGO bypass and is far more energetic than both 3$\alpha$ (5 orders of magnitude) and CAGO (10 orders of magnitude) at the conditions we're concerned with.  We can only get away with neglecting this rate because we don't expect protons/hydrogen to have any significant mass fraction.  This is a reasonable assumption for helium accretion but this rate will potentially be important in exploring scenarios where the helium layer is built up by burning hydrogen to helium depending on the amount of mixing that occurs.

I should note that W&K2011 state that

> The abundance of free protons can be significant because the composition of the fuel, almost entirely $^4$He and $^{14}$N, lacks any appreciable neutron excess.

My guess is that they're referring to the fact that such fuel will generate many free protons once $T > 1 \rm{GK}$ $\alpha$-chain burning is ignited.  In that case, we needn't worry as we're exploring the pre-explosive regime in which 3$\alpha$ dominates and the rest of the $\alpha$-chain is negligible.
