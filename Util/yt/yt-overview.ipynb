# volume rendering

```python
import yt.visualization.volume_rendering.api as vr
```

```python
c = np.array([0.0, 0.0, 0.0])
L = np.array([1.0, 1.0, 1.2])
W = 1.5*ds.domain_width
N = 720
north=[0.0,0.0,1.0]

# this is a hack at the moment -- all BCs are not supported yet
save_bc = ds.periodicity 
ds.periodicity = (True, True, True)
```

```python
field = ('boxlib', 'radial_velocity')
vals = [-1.e7, -5.e6, -2.5e6, 2.5e6, 5.e6, 1.e7]
sigma = 2.e5

cm = "coolwarm"

tf = vr.ColorTransferFunction((min(vals), max(vals)))
                    

for v in vals:
    tf.sample_colormap(v, sigma**2, colormap=cm)
```

```python
cam = vr.Camera(c, L, W, N, transfer_function=tf, ds=ds, 
                no_ghost=False, #data_source=reg,
                north_vector=north,
                fields = [field], log_fields = [False])
```

```python
im = cam.snapshot()
cam.draw_coordinate_vectors(im)
nim = cam.draw_domain(im)

# enhance contrast -- I need to make this happen automatically
max_val = im[:,:,:3].std() * 4.0
nim[:,:,:3] /= max_val

cam.save_annotated("my_fig.png", nim, dpi=200)
```

# surfaces

```python
# a hack to remove one zone in each direction to prevent boundaries from coming into play
root_dds = ds.domain_width/ds.domain_dimensions                              
half_w = ds.domain_width/2. - root_dds                                    
half_w[:] -= root_dds[:]                                                   
reg = ds.region(ds.domain_center, ds.domain_center-half_w, ds.domain_center+half_w)  

print reg
ds.periodicity = (True, True, True)
dd = ds.all_data()
surface = ds.surface(reg, "density", 7.e5)
```

```python
# color by radial velocity
colors = apply_colormap(surface["temperature"], cmap_name="hot")

colors = colors[0,:,:]/255.  # scale to [0,1]
colors[:,3] = 0.3            # alpha = 0.3
```

```python
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection


fig = figure()
ax = Axes3D(fig)

p3dc = Poly3DCollection(surface.triangles, linewidth=0.0)

p3dc.set_facecolors(colors)
ax.add_collection(p3dc)

max_extent = (surface.vertices.max(axis=1) - surface.vertices.min(axis=1)).max()
centers = (surface.vertices.max(axis=1) + surface.vertices.min(axis=1)) / 2
bounds = np.zeros([3,2])
bounds[:,0] = centers[:] - max_extent/2
bounds[:,1] = centers[:] + max_extent/2
ax.auto_scale_xyz(bounds[0,:], bounds[1,:], bounds[2,:])

show()
```
